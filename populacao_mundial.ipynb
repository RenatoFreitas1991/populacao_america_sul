```python
america_sul = dados.query('Continent == "South America"')
```

```python
display(america_sul)
```

```python
america_sul.query('CCA3 == "FLK"').index
```

```python
america_sul.drop(america_sul.query('CCA3 == "FLK"').index, axis=0, inplace=True)
```

```python
display(america_sul)
```

```python
america_sul.drop('Continent', axis=1, inplace=True)
```

```python
display(america_sul)
```

```python
america_sul.drop('Rank', axis=1, inplace=True)

display(america_sul)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.pie(america_sul['2020 Population'], labels=america_sul['Country/Territory'], autopct='%1.1f%%', startangle=90)
plt.title('População dos Países da América do Sul em 2020')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
selecao = america_sul['2020 Population'] > 20000000

america_sul_top_5 = america_sul[selecao]

display(america_sul_top_5)
```

```python
plt.figure(figsize=(10, 10))
plt.pie(america_sul_top_5['2020 Population'], labels=america_sul_top_5['Country/Territory'], autopct='%1.1f%%', startangle=90)
plt.title('População dos 5 Países da América do Sul mais populosos em 2020')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.pie(america_sul_top_5['2020 Population'], labels=None, autopct='%1.1f%%', startangle=90, pctdistance=0.85)
plt.title('População dos 5 Países da América do Sul mais populosos em 2020')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(america_sul_top_5['Country/Territory'], title="Países", loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
import numpy as np

# Get the list of countries from america_sul_top_5
countries = america_sul_top_5['Country/Territory'].tolist()

# Create a list of colors, with 'green' for Brazil and random colors for others
colors = []
for country in countries:
    if country == 'Brazil':
        colors.append('green')
    else:
        # Generate a random hex color
        colors.append('#%06X' % np.random.randint(0, 0xFFFFFF))

plt.figure(figsize=(12, 8))
plt.pie(america_sul_top_5['2020 Population'], labels=None, autopct='%1.1f%%', startangle=90, pctdistance=0.85, colors=colors)
plt.title('População dos 5 Países da América do Sul mais populosos em 2020')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(america_sul_top_5['Country/Territory'], title="Países", loc="center left", bbox_to_anchor=(1, 0.5))
plt.show()
```
